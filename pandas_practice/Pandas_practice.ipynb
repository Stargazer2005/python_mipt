````python
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
````

## 1. Создать DataFrame df c индексами из labels.



````python
df = pd.DataFrame(data,index=labels)
df
````

## 6. Вывести только те строки в которых количетсво посещений в ряде 'visits' больше 3.

````python
df[df.visits > 3]
````

## 11. Подсчитать сумму всех посещений.

````python
df['visits'].sum()
````

## 12. Найти средний возраст для каждого вида животного.

````python
df.groupby(['animal'], group_keys=True)[['age']].mean()
````

## 13. Добавьте любую мтроку k на ваш выбор и удалите ее.

````python
new_data = {'animal':['cat'], 'age':[1], 'visits':[2], 'priority':['no']}
new_labels = ['k']
new_df = pd.DataFrame(new_data,index=new_labels) 
df = pd.concat([df, new_df])
df
````

````python
df.drop('k')
````

## 14. Заменить в столбце priority параметр 'yes' и 'no' на True и False соответсвенно(Подсказка : метод .map)

````python
df['priority'] = df['priority'].map({'yes': True, 'no':False})
df
````

## 15. В столбце животные заменить 'snake' на 'python'(Подсказка : метод .replace)

````python
df['animal'] = df['animal'].replace('snake', 'python')
df
````

## 16. Для каждого вида животного и количества посещений найти средний возраст.

````python
df.groupby(['animal','visits'])[['age']].mean()
````

# Работа с файлами.

````python
titanik_passengers = pd.read_csv('titanic_data.csv', index_col='PassengerId')
titanik_passengers
````

## 1. Опишите данный датасет: какое расределение женщин/мужчин в нем? Сколько пассажиров ехало в каждом классе? Какой средний/минимальный/максимальный возраст пассажиров?

````python
print(titanik_passengers.groupby(['Sex']).size(),'\n')

print(titanik_passengers.groupby(['Pclass']).size(),'\n')

print('Mean age is: ',titanik_passengers['Age'].mean())

print('Min age is: ',titanik_passengers['Age'].min())

print('Max age is: ',titanik_passengers['Age'].max())
````

## 2. Сгруппируйте записи по классам пассажиров, в каждой группе посчитайте средний возраст. Используйте метод ``pandas.DataFrame.groupby``.

````python
print(titanik_passengers.groupby(['Pclass'])['Age'].mean(),'\n')
````

### 3. Слейте два датасета по колонке индекса.

````python
df_2 = pd.read_csv('titanic_surv.csv')
df_merge = pd.merge(titanik_passengers, df_2, on=df_2.index)
df_merge
````

### 4. Сколько всего выживших пассажиров? Выживших пассажиров по каждому из полов?

````python
print('Survived: ', df_merge['Survived'].sum(),'\n')
print(df_merge.groupby(['Sex'])['Survived'].sum())
````
